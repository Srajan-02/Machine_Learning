# K-Means Clustering

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset=pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:,[2,3,4]].values
```

```
print(x)
```

## Using the elbow method to find the optimal number of clusters

```
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11): #creating 10 k-means different objects
  kmeans = KMeans(n_clusters=i, init = 'k-means++',random_state=42) # to avoid the trap of random state we are initializing k-means++
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS')
plt.show()
```

## Training the K-Means model on the dataset

```
kmeans = KMeans(n_clusters=5, init = 'k-means++',random_state=42) # to avoid the trap of random state we are initializing k-means++
y_kmeans = kmeans.fit_predict(x)
```

```
print(y_kmeans)
```

## Visualising the clusters

selecting all the rows for first cluster i.e cluster 0.
first selecting for x and then for y,
0 for x and 1 for y

```
plt.scatter(x[y_kmeans == 0,0],x[y_kmeans == 0,1], s = 100, c = 'red',label = 'Cluster1')
plt.scatter(x[y_kmeans == 1,0],x[y_kmeans == 1,1], s = 100, c = 'blue',label = 'Cluster2')
plt.scatter(x[y_kmeans == 2,0],x[y_kmeans == 2,1], s = 100, c = 'green',label = 'Cluster3')
plt.scatter(x[y_kmeans == 3,0],x[y_kmeans == 3,1], s = 100, c = 'yellow',label = 'Cluster4')
plt.scatter(x[y_kmeans == 4,0],x[y_kmeans == 4,1], s = 100, c = 'pink',label = 'Cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s = 300, c = 'magenta', label = 'Centroids' )
plt.legend()
plt.title('Clusters of customer')
plt.xlabel('Annual income')
plt.ylabel('spending score')
plt.show()
```
